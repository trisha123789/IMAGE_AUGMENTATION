<b>Image augmentation</b>
 means artificially increasing your training dataset by applying transformations to images without changing their labels. It helps models generalize better and avoid overfitting.

<b>Why Image Augmentation is IMPORTANT</b>
<ul>
<li>Real-world data varies (lighting, angle, noise) </li>

<li>Deep models overfit small datasets</li>

<li>Makes model robust to unseen conditions</li>
<ul>
Think of it as: teaching the model reality.

<B> USING OPENCV </B>

```python
import cv2
image = cv2.imread('cat.jpg')
```

```python
image
```

<b>Using tensor(deep learning style)</b>

```python
import torch
img = torch.tensor(image) #this is mainly used during training
```

```python
img
```

```python
#(channel,height,width)
```

```python
from torchvision import transforms
transformed = transforms.RandomHorizontalFlip(p =1)
horizontal  = transformed(img)
plt.imshow(horizontal)
plt.axis("off")
plt.show() #use the image loaded by PIL
```

```python
from torchvision.transforms import functional as F
```

```python

flipped = F.hflip(img)
```

```python
plt.imshow(flipped)
plt.axis("off")
plt.show()
```

<b> VERTICAL FLIP </b>

```python
vflipped = F.vflip(img)
plt.imshow(vflipped)
plt.axis("off")
plt.show()
```

ROTATE

```python
rotated = F.rotate(img,angle=45)
plt.imshow(rotated)
plt.axis("off")
plt.show()
```

```python
brightness = F.adjust_brightness(img,brightness_factor=1.5)
plt.imshow(brightness)
plt.axis("off")
plt.show()
```

<ul>
<li> brighness_factor > 1 : brighter
</li>
<li> brightness_factor < 1 : darker
</li>
</ul>

```python
resized = F.resize(img, size=(128,128))

plt.imshow(resized)
plt.axis("off")
plt.title("Resized 128x128")
plt.show()
```

```python
gray = F.to_grayscale(img, num_output_channels=3)  # keep 3 channels for model

plt.imshow(gray)
plt.axis("off")
plt.title("Grayscale")
plt.show()
```

<b>

Adjust contrast
</b>

```python
contrast = F.adjust_contrast(img, contrast_factor=2.0)  # more contrast

plt.imshow(contrast)
plt.axis("off")
plt.title("High Contrast")
plt.show()
```
